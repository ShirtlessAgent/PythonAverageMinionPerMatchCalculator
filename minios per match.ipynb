```python
from tkinter import*
from bs4 import BeautifulSoup
import requests

import re





def webscrape():
         
    name = username.get()
    links = "https://lolprofile.net/summoner/eune/" + name +"#update"       
    atbilde = requests.get(links)
    doc = BeautifulSoup(atbilde.content, "html.parser")
    span = doc.find_all("span", class_="tooltip")
    i = 0
    str1 = "CS"
    minions = 0
    k = 0
    
    while True:
        try:
            index = span[i].string.find(str1)
            if (index != -1):
                skaits = int(re.search(r'\d+', span[i].string).group())
                int(skaits)
                minions = minions + skaits
                k += 1
            i += 1  
        except:
            minions = minions / k
            limited_float = round(minions, 2)
        
            
            minions2 = Label(window, text = "Average minions per game in last " + str(k) + " matches for " + str(name) + " has been " + str(limited_float))
            minions2.grid(row= 3, column=1)
            
            break            

window = Tk()
window.title("depresija")

username = Label(window, text = "username")
username.grid(row=0, column=0)

username = Entry(window)
username.grid(row=0, column = 1)

depresija = Button(window, text="depresija", command = webscrape)
depresija.grid(row=1, column=1)


euw = Button(window, text="euw", command = webscrape)
euw.grid(row=3, column=0)

eune = Button(window, text="eune", command = webscrape)
eune.grid(row=3, column=2)

window.geometry("500x500")
window.mainloop()





```
